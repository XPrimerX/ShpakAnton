# –ú–µ—Ç–∞: –∑–∞—Å–≤–æ—ó—Ç–∏ –≤–∞—Ä—ñ–∞–Ω—Ç–∏ —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—ó –∞–ª–≥–æ—Ä–∏—Ç–º—ñ–≤ –ø–æ—à—É–∫—É –∑–∞—Å–æ–±–∞–º–∏ Python —ñ –º–µ—Ç–æ–¥–∏ –æ—Ü—ñ–Ω—é–≤–∞–Ω–Ω—è —ó—Ö —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ.

## 1.–†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –ª—ñ–Ω—ñ–π–Ω–æ–≥–æ –ø–æ—à—É–∫—É –Ω–∞ Python

```python
# –Ø–∫—â–æ –µ–ª–µ–º–µ–Ω—Ç –Ω–µ –±—É–¥–µ –∑–Ω–∞–π–¥–µ–Ω–æ, –≤–∏–≤–æ–¥–∏—Ç—å—Å—è -1
def linear_search(a_list, x):
    i, length = 0, len(a_list)
    while i < length and x != a_list[i]:
        i += 1
    return i if i < length else -1
```

## 2.–û—Ü—ñ–Ω–∫–∞ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É –ª—ñ–Ω—ñ–π–Ω–æ–≥–æ –ø–æ—à—É–∫—É –≤ –û-–Ω–æ—Ç–∞—Ü—ñ—ó.

–ù–∞–π–≥—ñ—Ä—à–∏–π –≤–∏–ø–∞–¥–æ–∫: –µ–ª–µ–º–µ–Ω—Ç –≤—ñ–¥—Å—É—Ç–Ω—ñ–π —É –º–∞—Å–∏–≤—ñ –∞–±–æ –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –≤ –∫—ñ–Ω—Ü—ñ. –£ —Ü—å–æ–º—É –≤–∏–ø–∞–¥–∫—É –∞–ª–≥–æ—Ä–∏—Ç–º –ø–µ—Ä–µ–≤—ñ—Ä—è—î –≤—Å—ñ ùëõ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤, —Ç–æ–º—É –π–æ–≥–æ —á–∞—Å–æ–≤–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å O(n).
–ù–∞–π–∫—Ä–∞—â–∏–π –≤–∏–ø–∞–¥–æ–∫: —è–∫—â–æ –ø–æ—Ç—Ä—ñ–±–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –Ω–∞ –ø–µ—Ä—à—ñ–π –ø–æ–∑–∏—Ü—ñ—ó, —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –±—É–¥–µ O(1).

## 3.–ë—ñ–Ω–∞—Ä–Ω–∏–π –ø–æ—à—É–∫ –Ω–∞ Python

```python
def bin_search(a_list, x, left, right):
    if left > right or len(a_list) == 0:
        return -1
    middle = (left + right) // 2
    if a_list[middle] == x:
        return middle
    elif (a_list[middle] < x):
        return bin_search(a_list, x, middle + 1, right)
    else: # a_list[middle] > x
        return bin_search(a_list, x, left, middle - 1)
```

## 4.–æ—Ü—ñ–Ω–∫–∞ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É –±—ñ–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ—à—É–∫—É –≤ –û-–Ω–æ—Ç–∞—Ü—ñ—ó.

–ù–∞–π–≥—ñ—Ä—à–∏–π –≤–∏–ø–∞–¥–æ–∫:–°–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É ‚Äì O(log(n)), –¥–µ n ‚Äì –¥–æ–≤–∂–∏–Ω–∞ —Å–ø–∏—Å–∫—É –Ω–∞ –≤—Ö–æ–¥—ñ.

–ù–∞–π–∫—Ä–∞—â–∏–π –≤–∏–ø–∞–¥–æ–∫: O(1) —è–∫—â–æ —à—É–∫–∞–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –ø—Ä—è–º–æ –≤ —Ü–µ–Ω—Ç—Ä—ñ –º–∞—Å–∏–≤—É –ø—ñ—Å–ª—è –ø–µ—Ä—à–æ–≥–æ –ø–æ–¥—ñ–ª—É

## 5.–§—É–Ω–∫—Ü—ñ—è, —è–∫–∞ –ø–æ–≤–µ—Ä—Ç–∞—î –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –∑ —Ç–∞–∫–∏–º —Å–∫–ª–∞–¥–æ–º –ø–æ–ª—ñ–≤: (¬´n¬ª, ¬´time¬ª), –¥–µ n ‚Äì —Ä–æ–∑–º—ñ—Ä –º–∞—Å–∏–≤—É –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ—ó linear_searh();

```python
import time
import pandas as pd



def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1


def measure_linear_search(sizes):
    results = []
    for n in sizes:
        arr = list(range(n))  # –°—Ç–≤–æ—Ä—é—î–º–æ –≤–ø–æ—Ä—è–¥–∫–æ–≤–∞–Ω–∏–π –º–∞—Å–∏–≤
        target = n - 1  # –®—É–∫–∞–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç (–æ—Å—Ç–∞–Ω–Ω—ñ–π)

        start_time = time.time()
        linear_search(arr, target)
        end_time = time.time()

        results.append({"n": n, "time": end_time - start_time})

    return pd.DataFrame(results)

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–ª–∏–∫—É —Ñ—É–Ω–∫—Ü—ñ—ó
df = measure_linear_search([100, 1000, 5000, 10000, 50000])
print(df)
```

## 6.–§—É–Ω–∫—Ü—ñ—è, —è–∫–∞ –ø–æ–≤–µ—Ä—Ç–∞—î –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –∑ —Ç–∞–∫–∏–º —Å–∫–ª–∞–¥–æ–º–ø–æ–ª—ñ–≤: (¬´n¬ª, ¬´time¬ª), –¥–µ n ‚Äì —Ä–æ–∑–º—ñ—Ä –º–∞—Å–∏–≤—É –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ—ó bin_searh();

```python
import time
import pandas as pd

def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1


def measure_binary_search(sizes):
    results = []
    for n in sizes:
        arr = list(range(n))  # –°—Ç–≤–æ—Ä—é—î–º–æ –≤–ø–æ—Ä—è–¥–∫–æ–≤–∞–Ω–∏–π –º–∞—Å–∏–≤
        target = n - 1  # –®—É–∫–∞–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç (–æ—Å—Ç–∞–Ω–Ω—ñ–π)

        start_time = time.time()
        binary_search(arr, target)
        end_time = time.time()

        results.append({"n": n, "time": end_time - start_time})

    return pd.DataFrame(results)


# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–ª–∏–∫—É —Ñ—É–Ω–∫—Ü—ñ—ó
df = measure_binary_search([100, 1000, 5000, 10000, 50000])
print(df)
```
