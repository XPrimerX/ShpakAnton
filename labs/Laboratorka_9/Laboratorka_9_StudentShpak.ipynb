## 1.–í–ª–∞—Å–Ω–∞ —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—è Z-—Ñ—É–Ω–∫—Ü—ñ—ó:

```python
def compute_z_array(s: str) -> list[int]:
    n = len(s)
    z = [0] * n
    z[0] = n  # –ü–æ–≤–Ω–∞ –¥–æ–≤–∂–∏–Ω–∞ ‚Äî –ø–µ—Ä—à–∏–π –µ–ª–µ–º–µ–Ω—Ç

    left, right = 0, 0
    for i in range(1, n):
        if i <= right:
            z[i] = min(right - i + 1, z[i - left])
        while i + z[i] < n and s[z[i]] == s[i + z[i]]:
            z[i] += 1
        if i + z[i] - 1 > right:
            left, right = i, i + z[i] - 1
    return z
```

## 2.–í–ª–∞—Å–Ω–∞ —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –ø–æ—à—É–∫—É —à–∞–±–ª–æ–Ω—É —É —Ç–µ–∫—Å—Ç—ñ —á–µ—Ä–µ–∑ Z-—Ñ—É–Ω–∫—Ü—ñ—é:

```python
def find_substring(pattern: str, text: str) -> list[int]:
    combined = pattern + "#" + text
    z = compute_z_array(combined)
    positions = []

    pattern_length = len(pattern)
    for i in range(pattern_length + 1, len(combined)):
        if z[i] == pattern_length:
            positions.append(i - pattern_length - 1)
    return positions
```

## 3.–í–ª–∞—Å–Ω–∞ —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—è —Å—Ç–∏—Å–Ω–µ–Ω–Ω—è —Ä—è–¥–∫–∞ –Ω–∞ –æ—Å–Ω–æ–≤—ñ –ø–µ—Ä—ñ–æ–¥–∏—á–Ω–æ—Å—Ç—ñ —á–µ—Ä–µ–∑ Z-—Ñ—É–Ω–∫—Ü—ñ—é:

```python
def compress_string_by_z(s: str) -> str:
    z = compute_z_array(s)
    for i in range(1, len(s)):
        if z[i] + i == len(s) and len(s) % i == 0:
            return s[:i]
    return s
```
