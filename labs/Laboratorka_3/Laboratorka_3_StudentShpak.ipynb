## 1.Виконав теоретичні розрахунки асимптотичної складності алгоритмів сортування методами вставляння та бульбашки.
1. Сортування вставками (Insertion Sort)
Найгірший випадок (O(n²)): коли масив спочатку відсортований у зворотному порядку. 
Найкращий випадок (O(n)): коли масив уже відсортований, і кожен елемент просто перевіряється, не потребуючи перестановок.

2. Сортування бульбашкою (Bubble Sort)
Найгірший випадок (O(n²)): якщо масив відсортований у зворотному порядку, кожен елемент переміщається на своє місце через численні перестановки.
Найкращий випадок (O(n)): якщо масив уже відсортований.

## 2.Реалізація на Python алгоритм сортування методом бульбашки.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
```

## 3.Дослідження реальний час виконання обох алгоритмів на Python Tinsert(n) та Tbuble(n):

```python
import time
import random
import matplotlib.pyplot as plt

# Алгоритм сортування методом бульбашки
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

# Алгоритм сортування методом вставок
def insertion_sort(nums):
    # починаємо з другого елемента
    for j in range(1, len(nums)):
        key = nums[j]
        # Зберігаємо посилання на індекс першого елемента
        i = j - 1
        # Елементи відсортованого сегрмента переміщуємо вперед, якщо вони більше елемента вставки
        while i >= 0 and nums[i] > key:
            nums[i + 1] = nums[i]
            i -= 1
        # Вставляємо елемент
        nums[i + 1] = key

# Функція вимірювання часу виконання
def measure_time(sort_func, arr):
    start_time = time.time()
    sort_func(arr.copy())  # Копія масиву, щоб уникнути змін
    return time.time() - start_time

# Розміри масивів для тестування
sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000]

# Збереження часу виконання
bubble_times = []
insertion_times = []

for size in sizes:
    test_array = [random.randint(0, 10000) for _ in range(size)]
    bubble_times.append(measure_time(bubble_sort, test_array))
    insertion_times.append(measure_time(insertion_sort, test_array))

# Побудова графіків
plt.figure(figsize=(10, 6))
plt.plot(sizes, bubble_times, label='Bubble Sort', marker='o')
plt.plot(sizes, insertion_times, label='Insertion Sort', marker='s')
plt.xlabel('Розмір масиву (n)')
plt.ylabel('Час виконання (секунди)')
plt.title('Залежність часу сортування від розміру масиву')
plt.legend()
plt.grid()
plt.show()

```

## 5.Відповіді на запитання:

1. Що таке O-нотація і чим вона відрізняється від Ω -нотації?
O-нотація: Показує максимальну складність алгоритму (найгірший випадок).
Ω-нотація: Показує мінімальну складність алгоритму (найкращий випадок).
Вони відображають верхню та нижню межі зростання часу виконання або використання ресурсів.

2. Яку часову складність має алгоритм сортування методом вставляння за найсприятливіших умов, коли масив відсортовано?
Коли масив уже відсортовано, часову складність алгоритму сортування методом вставляння становить Ω(𝑛). У цьому випадку кожен елемент перевіряється лише один раз, і не відбувається переміщення даних. Таким чином, сортування здійснюється лінійно.

3. Визначення ефективного алгоритму
Ефективний алгоритм — це алгоритм, який вирішує завдання за прийнятний час і з оптимальним використанням ресурсів, таких як пам'ять або процесорний час. Його ефективність зазвичай оцінюється за допомогою часової та просторової складності.

4. Пояснити поняття головного параметра (розміру) завдання.
Головний параметр завдання — це міра або характеристика, яка відображає обсяг вхідних даних. Наприклад:
У сортуванні розмір завдання — це кількість елементів у масиві (𝑁).
У роботі з графами розмір завдання може бути числом вершин або ребер.

5. Функція часової складності має вигляд: F(N) = $N^3$ + $7N^2$ − 14N. Як записати асимптотичну складність у O-нотації?
Для визначення асимптотичної складності використовуємо найшвидше зростаючий член, ігноруючи коефіцієнти та менш значущі члени. Найшвидше зростаючий член тут — $𝑁^3$
Тому асимптотична складність у O-нотації записується як 𝑂($𝑁^3$).
