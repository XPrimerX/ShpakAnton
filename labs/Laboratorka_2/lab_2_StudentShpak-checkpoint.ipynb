## Відповіді на запитання:
1. Що таке pip і conda та чим вони відрізняються між собою?
pip встановлює пакети лише для Python, а conda може встановлювати також бібліотеки для інших мов.
pip бере пакети з PyPI, а conda – з власних репозиторіїв.
conda керує залежностями більш ефективно, ніж pip, що зменшує ризик конфліктів між пакетами.

3.Написати процедуру на Python сортування масиву методом бульбашки.


```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:  
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
```
